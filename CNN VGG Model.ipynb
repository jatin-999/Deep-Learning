```python
import os
import numpy as np
import cv2
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
```

```python
img_width, img_height = 224, 224
num_classes = 2
epochs = 5
batch_size = 32

# Load images and labels
images = []
labels = []
```

```python
image_dir = "C:\\Users\91992\Desktop\Deep Learning Project\Images 2"
for filename in os.listdir(image_dir):
    if filename.endswith(".jpeg"):
        image = cv2.imread(os.path.join(image_dir, filename))
        image = cv2.resize(image, (img_width, img_height))
        images.append(image)
        
        # Assuming all images are labeled as 1 (rip tide) for now
        labels.append(1) 
```

```python
images = np.array(images)
labels = np.array(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.1, random_state=1)
```

```python
X_train = preprocess_input(X_train)
X_test = preprocess_input(X_test)
```

```python
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_width, img_height, 3))
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu'),
    Dense(num_classes, activation='softmax')
])
```

```python
model.compile(optimizer=Adam(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(X_test, y_test))
```
